# Load and Plot GeoTiff

This notebook simply shows how a GeoTiff can be loaded and displayed with Python - including adding contour lines and some other tricks.

The most difficult part is getting the appropriate geographic libraries installed... this can be daunting, especially on windows. However, with new versions of Anaconda, this should work now (using `conda`)

```python
import gdal, osr
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Open raster data
dome_sub = gdal.Open("../data/dome_sub_sub_utm.tif")
```

```python
arr_sub = dome_sub.GetRasterBand(1).ReadAsArray()
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(5,3))
# plt.imshow(arr[450:500,450:500], origin='lower left', cmap='viridis', interpolation='nearest')
plt.imshow(arr_sub, # origin='lower left', 
           cmap='viridis', 
           interpolation='nearest')
```
